## Sentiment table

```python
sentiment_col = [i for i in df_result.columns if 'sentiment' in i]
sentiment_col
```

```python
df_sentiment = pd.merge(df_result[sentiment_col+['review_id']], df_review[['review_id', 'date', 'biz_id']], on='review_id').merge(df_business[['biz_name','biz_id','biz_address','biz_city','biz_state', 'biz_lat', 'biz_lng']], on='biz_id')
df_sentiment.head()
```

```python
df_sentiment.groupby('biz_name')['review_id'].count().sort_values(ascending=False).index[0]
```

```python
df_sentiment.groupby('biz_id')['review_id'].count().sort_values(ascending=False).index[0:10]
```

```python
top10 = df_sentiment.groupby('biz_name')['review_id'].count().sort_values(ascending=False).index[0:10]
```

```python
top10
```

```python
df_top10 = df_sentiment[df_sentiment['biz_name'].isin(top10)]
```

```python
df_top10.head()
```

```python
import datetime
import calendar

def week_of_month(tgtdate):
#     tgtdate = tgtdate.to_datetime()

    days_this_month = calendar.mdays[tgtdate.month]
    for i in range(1, days_this_month):
        d = datetime.datetime(tgtdate.year, tgtdate.month, i)
        if d.day - d.weekday() > 0:
            startdate = d
            break
    # now we canuse the modulo 7 appraoch
    return (tgtdate - startdate).days //7 + 1

df_top10['date'] = pd.to_datetime(df_top10['date'])
df_top10['week_of_month'] = df_top10['date'].apply(week_of_month)
df_top10['week_of_year'] = df_top10['date'].dt.week
df_top10['month'] = df_top10['date'].dt.month
df_top10['year'] = df_top10['date'].dt.year
df_top10['dates'] = df_top10['date'].dt.date
```

```python
df_top10.head()
```

```python
df_top10.columns
```

```python
df_top10['cleanliness_sentiment'] = df_top10['cleanliness_sentiment'].astype(float).round(2)
df_top10['food_sentiment'] = df_top10['food_sentiment'].astype(float).round(2)
df_top10['location_sentiment'] = df_top10['location_sentiment'].astype(float).round(2)
df_top10['service_sentiment'] = df_top10['service_sentiment'].astype(float).round(2)
df_top10['price_sentiment'] = df_top10['price_sentiment'].astype(float).round(2)
```

```python
df_top10.fillna(0, inplace=True)
```

```python
df_top10.rename(columns={'biz_lat':'Latitude', 'biz_lng':'Longitude'}, inplace=True)
```

# Emotion table

```python
emotion_col = ['cleanliness_anger', 'cleanliness_disgust', 'cleanliness_fear','cleanliness_joy', 'cleanliness_sadness',
       'food_anger', 'food_disgust', 'food_fear', 'food_joy', 'food_sadness', 'location_anger', 'location_disgust', 'location_fear',
       'location_joy', 'location_sadness', 'price_anger', 'price_disgust', 'price_fear', 'price_joy', 'price_sadness', 'service_anger', 'service_disgust',
       'service_fear', 'service_joy', 'service_sadness']
emotion_col
```

```python
df_emotion = pd.merge(df_result[emotion_col+['review_id']], df_review[['review_id', 'date', 'biz_id']], on='review_id').merge(df_business[['biz_name','biz_id','biz_address','biz_city','biz_city','biz_state', 'biz_lat', 'biz_lng']], on='biz_id')
df_emotion.head()
```

```python
df_emotion_top10 = df_emotion[df_emotion['biz_name'].isin(top10)]
df_emotion_top10.head()
```

```python
len(df_emotion_top10)
```

```python
df_cleanliness = pd.DataFrame()
df_food = pd.DataFrame()
df_location = pd.DataFrame()
df_price = pd.DataFrame()
df_service = pd.DataFrame()
```

```python
from tqdm import tqdm
```

```python
for aspect_emotion in tqdm(emotion_col):
    aspect = aspect_emotion.split('_')[0]
    emotion = aspect_emotion.split('_')[1]
    if 'cleanliness' in aspect:
        df_temp = df_emotion_top10[['review_id',aspect_emotion]]
        df_temp.rename(columns={aspect_emotion:'cleanliness'}, inplace=True)
        df_temp['emotion'] = emotion
        df_cleanliness = df_cleanliness.append(df_temp)
    elif 'food' in aspect:
        df_temp = df_emotion_top10[['review_id',aspect_emotion]]
        df_temp.rename(columns={aspect_emotion:'food'}, inplace=True)
        df_temp['emotion'] = emotion
        df_food = df_food.append(df_temp)
    elif 'location' in aspect:
        df_temp = df_emotion_top10[['review_id',aspect_emotion]]
        df_temp.rename(columns={aspect_emotion:'location'}, inplace=True)
        df_temp['emotion'] = emotion
        df_location = df_location.append(df_temp)
    elif 'price' in aspect:
        df_temp = df_emotion_top10[['review_id',aspect_emotion]]
        df_temp.rename(columns={aspect_emotion:'price'}, inplace=True)
        df_temp['emotion'] = emotion
        df_price = df_price.append(df_temp)
    elif 'service' in aspect:
        df_temp = df_emotion_top10[['review_id',aspect_emotion]]
        df_temp.rename(columns={aspect_emotion:'service'}, inplace=True)
        df_temp['emotion'] = emotion
        df_service = df_service.append(df_temp)
```

```python
df_combine = pd.merge(df_cleanliness, df_food, on=['review_id','emotion']).merge(df_location, on=['review_id','emotion']).merge(df_service, on=['review_id','emotion']).merge(df_price, on=['review_id','emotion'])
```

```python
df_combine = pd.merge(df_combine, df_review[['review_id', 'date', 'biz_id']], on='review_id').merge(df_business[['biz_name','biz_id','biz_address','biz_city','biz_state', 'biz_lat', 'biz_lng']], on='biz_id')
df_combine.head()
```
